```python
import pandas as pd
```

```python
file_path = "../data/raw/DESTATIS-45212BH002 time series.csv"
df = pd.read_csv(file_path)
```

```python
selected_columns = [
    "period",
    "Germany – Motor trade and retail trade – at current prices – Turnover (DESTATIS/45212BH002/DG.WZ08-45-01.NOMINAL.UMS002)",
    "Germany – Motor trade and retail trade – at constant prices – Turnover (DESTATIS/45212BH002/DG.WZ08-45-01.REAL.UMS002)"
]
```

```python
for col in df.columns:
    print(col)
```

```python
retail_df = df[selected_columns]
```

```python
retail_df.head()
```

```python
retail_df = retail_df.rename(
    columns={
        selected_columns[1]: "turnover_nominal",
        selected_columns[2]: "turnover_real"
    }
)
```

```python
retail_df.head()
```

```python
retail_df["period"].head(6)
```

```python
def convert_period_to_date(period):
    year, half = period.split("-")

    if half == "S1":
        return f"{year}-01-01"
    elif half == "S2":
        return f"{year}-07-01"
    else:
        return None
```

```python
retail_df["date"] = retail_df["period"].apply(convert_period_to_date)
```

```python
retail_df["date"] = pd.to_datetime(retail_df["date"])
```

```python
retail_df[["period", "date"]].head(6)
```

```python
retail_df.isna().sum()
```

```python
retail_df[["date", "turnover_nominal", "turnover_real"]].head(15)
```

```python
retail_df.dropna(subset=["turnover_nominal", "turnover_real"], inplace=True)
```

```python
retail_df.head()
```

```python
retail_df.isna().sum()
```

```python
retail_df = retail_df.set_index("date")
```

```python
retail_df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
retail_df["inflation_gap"] = retail_df["turnover_nominal"] - retail_df["turnover_real"]
plt.figure(figsize=(12, 6))

plt.plot(retail_df.index, retail_df["turnover_nominal"], label="Nominal Turnover")
plt.plot(retail_df.index, retail_df["turnover_real"], label="Real Turnover")

plt.title("German Retail Turnover: Nominal vs Real")
plt.xlabel("Year")
plt.ylabel("Turnover Index")
plt.legend()
plt.show()
```

 Nominal turnover represents retail sales measured at current prices, while real turnover adjusts for inflation and reflects actual business activity.

From the plot, we observe periods where nominal turnover continues to rise while real turnover grows more slowly or remains flat. This indicates that reported revenue growth during these periods is primarily driven by price increases rather than an increase in the volume of goods sold.


For retail businesses, this can signal underlying demand weakness and potential pressure on profitability, as higher prices do not necessarily translate into stronger real performance.
